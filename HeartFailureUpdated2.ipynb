### People having heart diseases

```python
heartDiseases_df=dataset_df["HeartDisease"]
```

```python
print(heartDiseases_df)
```

```python
hdtrue_df=dataset_df[dataset_df["HeartDisease"] == 1]
hdfalse_df=dataset_df[dataset_df["HeartDisease"] == 0]
```

```python
print(hdtrue_df)
```

```python
totalPatients=len(heartDiseases_df)
```

```python
print(totalPatients)
```

```python
heartPatients=len(hdtrue_df)
```

```python
print(heartPatients)
```

```python
percentageOfHeartPatients=(heartPatients/totalPatients) * 100
```

```python
print(percentageOfHeartPatients)
```

```python
y=[len(dataset_df[dataset_df["HeartDisease"]==1]), len(dataset_df[dataset_df["HeartDisease"]==0])]
x=['Have Heart Disease', 'Dont Have Heart Disease' ]
plt.bar(x,y)
plt.title("Heart Patients vs Non Heart Patients")
```

```python
plt.pie([len(dataset_df[dataset_df["HeartDisease"]==1]), len(dataset_df[dataset_df["HeartDisease"]==0])])
plt.title("total patient distribution with reference to heart disease")
plt.legend(["Have Heart disease","dont have heart disease"])
```

# impact of gender

## Females

```python
females_df=dataset_df[dataset_df["Sex"]== 'F']
```

```python
femaleHeartPatients=females_df[females_df["HeartDisease"]==1]
```

```python
totalFemales=len(females_df)
```

```python
print(totalFemales)
```

```python
femalePatients=len(femaleHeartPatients)
```

```python
print(femalePatients)
```

```python
femalePercentage = femalePatients / totalFemales * 100
```

```python
print(femalePercentage)
```

```python
plt.pie([femalePercentage,(100-femalePercentage)])
plt.title("Heart Disease among female patients")
plt.legend(["Have Heart Disease"," Dont Have Heart Disease"])
```
